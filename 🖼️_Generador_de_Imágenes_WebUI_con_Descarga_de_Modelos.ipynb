```python
# ==========================================
#@markdown ### ğŸ“¥âœ¨ **DESCARGA DE VAE, EMBEDDINGS Y UPSCALERS** âœ¨ğŸ“¥
# ==========================================

# ğŸ¨ **Links para VAE:**
vae_links = "https://civitai.com/api/download/models/290640?type=VAE&format=SafeTensor"  #@param {type:"string"}

# ğŸ§  **Links para Embeddings:**
embeddings_links = ""  #@param {type:"string"}

# ğŸ” **Links para Upscalers:**
upscalers_links = ""  #@param {type:"string"}

#@markdown ### â¬‡ï¸ **Descarga de Modelos y LoRAs**


# ğŸ­ **Links para Checkpoint:**
checkpoint_links = "https://civitai.com/api/download/models/1490781?type=Model&format=SafeTensor&size=pruned&fp=fp16"  #@param {type:"string"}

# ğŸ–Œï¸ **Links para LoRA:**
lora_links = "https://civitai.com/api/download/models/1558670?type=Model&format=SafeTensor"  #@param {type:"string"}

import os

def download_links(link_string, output_path):
    """
 #@markdown #####   ğŸ“¡ Descarga archivos desde una lista de links separados por comas: (Url_1, Url_2, etc...)
    Args:
        link_string (str): Links separados por comas.
        output_path (str): Directorio donde guardar los archivos.
    """
    links = [link.strip() for link in link_string.split(",") if link.strip()]
    os.makedirs(output_path, exist_ok=True)  # âœ… Asegura que el directorio exista

    for link in links:
        print(f"â¬‡ï¸ Descargando: {link} a {output_path}...")
        try:
            get_ipython().run_line_magic('download', link)
            print(f"âœ… Descargado con Ã©xito: {link}")
        except Exception as e:
            print(f"âŒ Error al descargar {link}: {str(e)}")

# ğŸš€ **Iniciando descargas...**
print("\nğŸ”µ Descargando **VAE**...")
%cd -q $VAE
download_links(vae_links, "./VAE")

print("\nğŸŸ¢ Descargando **Embeddings**...")
%cd -q $Embeddings
download_links(embeddings_links, "./Embeddings")

print("\nğŸŸ  Descargando **Upscalers**...")
%cd -q $Upscalers
download_links(upscalers_links, "./Upscalers")

print("\nğŸ‰ Â¡Descargas completadas! ğŸ”¥")

# ğŸš€ **Iniciando descargas...**
print("''' Checkpoint '''")
download_links(checkpoint_links, "./CKPT")

print("''' LoRA '''")
download_links(lora_links, "./LORA")

print("\nâœ… Descarga completada. Â¡Listo para usar los modelos en Stable Diffusion! ğŸš€")

# ========================================
#@markdown ##### ğŸ”Œâœ¨ **DESCARGA Otras EXTENSIONES (Github Repositories)** âœ¨ğŸ”Œ
# ========================================

# ğŸ“¥ **Links para Extensiones (Git Repositories)** ğŸ”—
# (Separar por comas y un espacio si son mÃºltiples)
extensions_links = "https://github.com/yankooliveira/sd-webui-photopea-embed"  #@param {type:"string"}

import os

def git_clone_links(link_string, output_path):
    """
#@markdown #####    ğŸ”„ Clona repositorios Git desde una lista de links separados por comas.
    Args:
        link_string (str): Links separados por comas.
        output_path (str): Directorio donde clonar los repositorios.
    """
    links = [link.strip() for link in link_string.split(",") if link.strip()]
    os.makedirs(output_path, exist_ok=True)  # âœ… Asegura que el directorio exista

    for link in links:
        print(f"ğŸš€ Clonando: {link} a {output_path}...")
        try:
            os.system(f"git -C {output_path} clone {link}")
            print(f"âœ… ClonaciÃ³n completada: {link}")
        except Exception as e:
            print(f"âŒ Error al clonar {link}: {str(e)}")

# ğŸ“‚ **Descargando extensiones...**
print("''' ğŸ”§ Extensiones '''")
%cd -q $Extensions
git_clone_links(extensions_links, "./")

print("ğŸ‰ Â¡Extensiones instaladas con Ã©xito! ğŸ”¥")
```
